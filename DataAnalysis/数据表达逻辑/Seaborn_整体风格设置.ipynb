```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# 创建正弦函数 及 图表
def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x + i * 1.5) * (7-i) * flip)

# 绘图
sinplot()
```

```python
sns.set() # 设置默认风格
sinplot()
plt.grid(linestyle='--')
```

```python
# 2.set_style()
# 切换seaborn 风格 包括(white,dark,whitegrid,darkgrid,ticks)

fig = plt.figure(figsize=(6,6))

ax1 = fig.add_subplot(2,1,1)
sns.set_style('whitegrid')
data = np.random.normal(size=(20,6)) + np.arange(6)/2
sns.boxplot(data=data)
plt.title('style - whitegrid')  # 这里任然可以使用matplotlib参数

ax2 = fig.add_subplot(2,1,2)
sinplot()
```

```python
# 3、despine()
# 设置图表坐标轴
# seaborn.despine(fig=None,ax=None,top=True,right=True,left=False,bottom=False,offset=None,trim=False)

sns.set_style('ticks') # 设置整体风格

fig = plt.figure(figsize=(6,9))
plt.subplots_adjust(hspace=0.3)
# 创建图表

ax1 = fig.add_subplot(3,1,1)
sinplot()
sns.despine()
# 删除了 上、右坐标轴

ax2 = fig.add_subplot(3,1,2)
sns.violinplot(data=data)
# sns.despine(offset=10,trim=True)
# offset: 与坐标轴之间的偏移量
# trim: 为True时，将坐标轴限制在数据 最大最小值

ax3 = fig.add_subplot(3,1,3)
sns.boxplot(data=data,palette='deep')
# sns.despine(left=True,right=False)
# top,right,left,bottom:布尔型，为True时不显示
```

```python
# 4.axes_style()
# 设置局部图表风格，可以和with配合的用法

with sns.axes_style('darkgrid'):
    plt.subplot(211)
    sinplot()
# 设置局部图表风格，用with做代码块区分

sns.set_style('whitegrid')
plt.subplot(212)
sinplot()
```

```python
# 5.set_context()
# 设置显示比例尺度
# 选择包括 ： paper,notebook,talk,poster
sns.set_context('notebook')
sinplot()  # 默认为notebook
```
