```python
'''
【2018年上半年电影数据】
    数据清洗： 保留 制片国家 、豆瓣评分、电影名称、电影类型 四个字段
    分析：1.那个国家的在7分以上的电影最多
          2. 分别是什么电影

'''
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore') # 不发出警报

# 解决显示中文的问题
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
# 调整当前的工作路径
import os 
os.chdir('E:\BaiduNetdiskDownload')
print(os.getcwd())

# 读取文件数据
data = pd.read_excel('电影资源整理180406版.xlsx',sheetname='影视资源',header=1)

data2 = data[['制片国家','豆瓣评分','影视名称','影视形式']]
data2 = data2.dropna()
data2.head()
```

```python
# 自定义函数 清理 ‘豆瓣评分’ = ‘未知’ 项
def replace_nd(df,*args):
    data_item = []
    for i in args:
        for j in df[i]:
            if j == '未知':
                j = 5.0
            data_item.append(j)
    return data_item
    

data2['豆瓣评分'] = replace_nd(data2,'豆瓣评分')
```

```python
# 各国2018年上半年电影产出
data2['制片国家'] = data2['制片国家'].str.strip()
country_count = data2.groupby('制片国家').count()
country_count.columns = ['电影个数1','电影个数2','电影个数3']
result = country_count['电影个数1']

x = result.index
y = result
plt.figure(figsize=(40,8))
plt.bar(x,y,width=0.5,facecolor='lightskyblue',edgecolor='white',yerr=y)
plt.grid(linestyle='--')
plt.xticks(fontsize=18,color='white')
plt.yticks(fontsize=18,color='white')
```

```python
# 计算出 评分大于 7 分的电影 并且新建列“佳片”
rule = data2['豆瓣评分']
result = []
for i in rule:
    n = float(i)
    result.append(n)
data2['豆瓣评分'] = result
data2['佳片'] = data2['豆瓣评分']>7.0
data2.head()
```

```python
#  根据 佳片 和 制片国家分类
niceMovie_count = data2.groupby(['佳片','制片国家']).count()
# niceMovie_count.columns
del niceMovie_count['影视名称']
del niceMovie_count['影视形式']
# 重新划分组
niceMovie_count.columns = ['影片个数']
niceMovie_count.reset_index(inplace=True) # 把分组的条件转换成列名
niceMovie_count.head()
```

```python
# 分组后的结果
F_Movie = niceMovie_count.groupby('佳片').get_group(False)[['制片国家','影片个数']]
T_Movie = niceMovie_count.groupby('佳片').get_group(True)[['制片国家','影片个数']]

plt.figure(figsize=(30,4))
x1 = T_Movie['制片国家']
y1 = T_Movie['影片个数']

x2 = F_Movie['制片国家']
y2 = F_Movie['影片个数']

# plt.bar(x1,y1,width=0.5,facecolor='yellowgreen',edgecolor='white',yerr=y1)
plt.bar(x2,y2,width=0.5,facecolor='lightskyblue',edgecolor='white',yerr=y2)
plt.grid(linestyle='--')
plt.title('地域(国家)2018年上半年烂片数量',fontsize=18,color='white')
plt.xticks(fontsize=18,color='white')
plt.yticks(fontsize=18,color='white')

plt.figure(figsize=(30,4))
plt.bar(x1,y1,width=0.5,facecolor='yellowgreen',edgecolor='white',yerr=y1)
plt.grid(linestyle='--')
plt.title('地域(国家)2018年上半佳片数量',fontsize=18,color='white')
plt.xticks(fontsize=18,color='white')
plt.yticks(fontsize=18,color='white')
```

```python
# 自定义方法，清洗横坐标
def unitize_x(df,*args):
    F_Movie = df.groupby('佳片').get_group(False)[['制片国家','影片个数']]
    T_Movie = df.groupby('佳片').get_group(True)[['制片国家','影片个数']]
    

    Movie_and = set(F_Movie['制片国家']) & set(T_Movie['制片国家'])
    Movie_or = set(F_Movie['制片国家']) | set(T_Movie['制片国家'])
    Movie = list(Movie_or - Movie_and)
    
    data11 = {'制片国家':Movie,'影片个数':list(np.random.rand(len(Movie))*0)}
    s1 = pd.DataFrame(data11)

    pp_T = pd.merge(T_Movie,s1,left_on=['制片国家'],right_on=['制片国家'],how='outer',sort=True)
    pp_T.fillna(value=0,inplace=True)
    
    pp_F = pd.merge(F_Movie,s1,left_on=['制片国家'],right_on=['制片国家'],how='outer',sort=True)
    pp_F.fillna(value=0,inplace=True)
    
    return pp_T[['制片国家','影片个数_x']],pp_F[['制片国家','影片个数_x']]

# 绘图前准备
T_movie,F_movie = unitize_x(niceMovie_count)

x = T_movie['制片国家']
y1 = T_movie['影片个数_x']
y2 = -F_movie['影片个数_x']

# 配置刻度
# from matplotlib.ticker import MultipleLocator,FormatStrFormatter

# ax = plt.subplot(111)
# ymajorLocator   = MultipleLocator(5) # 将x轴次刻度标签设置为5的倍数 


# 开始绘图
plt.figure(figsize=(35,8))
plt.bar(x,y1,width=0.5,facecolor='yellowgreen',edgecolor='white',yerr=y1)
plt.bar(x,y2,width=0.5,facecolor='lightskyblue',edgecolor='white',yerr=y2)
plt.grid(linestyle='--')
plt.title('地域(国家)2018年上半年(佳片/烂片)数量',fontsize=18,color='white')
plt.xticks(fontsize=18,color='white')
plt.yticks(fontsize=18,color='white') 
```

```python
# 计算出地域(国家)的 佳片/烂片占比
T_sum = sum(T_movie['影片个数_x'])
F_sum = sum(F_movie['影片个数_x'])
T_rate = T_movie['影片个数_x'] / T_sum * 100
F_rate = F_movie['影片个数_x'] / F_sum * 100

plt.axis('equal')
plt.pie(
    T_rate,
   labels = T_rate.index,
   autopct='%.2f%%',
   pctdistance=0.6,
   labeldistance = 1.2,
   shadow = True,
   startangle=0,
   radius=1.5,
   frame=False
)
```
