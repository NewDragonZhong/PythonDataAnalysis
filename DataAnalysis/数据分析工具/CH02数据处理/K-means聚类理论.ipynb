```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 创建数据
from sklearn.datasets.samples_generator import make_blobs
# make_blobs 聚类数据生成器

x,y_true = make_blobs(n_samples=300,centers=4,cluster_std=0.5,random_state=0)
# 生成300条数据 ，分为4类数据，方差一致都为0.5
# n_features: 每个样本的特征数
# random_state: 随机数种子
# x : 生成数据值 ，y: 生成数据对应的类别标签

# 绘图 s:(我猜的是按比例缩放)
plt.scatter(x[:,0],x[:,1],s=10,alpha=0.8)
plt.grid()
```

```python
# 导入模块
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4) # 这里为4簇
kmeans.fit(x)
y_kmeans = kmeans.predict(x)
centroids = kmeans.cluster_centers_
# 构建模型，并预测出样本的类别y_kmeans
# kmeans.cluster_centers_ : 得到不同簇的中心点

plt.scatter(x[:,0],x[:,1],c=y_kmeans,cmap='Dark2',s=50,marker='x',alpha=0.5)
plt.scatter(centroids[:,0],centroids[:,1],c=[0,1,2,3],cmap='Dark2',marker='o')
plt.title('K-means 300 points\n')
plt.xlabel('Value1')
plt.ylabel('Value2')
plt.grid()
```
