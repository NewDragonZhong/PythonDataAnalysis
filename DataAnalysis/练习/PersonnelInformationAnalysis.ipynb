```python
'''
    1.按单位(人)分析
        * Top10排行榜
            -- 加班/剩余假期(剩余假期+剩余调休)/工作延时次数/司龄/缺勤次数/迟到次数/平均工时
            -- 十大最忙员工？(公式指定原则：按照待在公司的时间长短来确定)
                * 计算公式：y = 剩余假期*0.1+加班*0.2+工作延时次数*0.3+平均工时*0.4 - (缺勤次数+迟到早退次数)*0.5 
'''
```

```python
# 读取文件_将此变量作为一个 内存变量保存
read_me = pd.read_excel('人员信息.xlsx',sheetname='sheet1',header=0)
# 定义一个函数排序字段
def data_ETL(data,field='公休日加班',reject='次',num=10):
    # 清洗数据并转换数据类型
    if data[field].dtype == 'object':
        data[field] = data[field].str.replace(reject,'').astype(np.float64)
    data = data.sort_values([field],ascending=False)
    return data[[field,'姓名']].head(num)
```

```python
# 加班TOP10
overtime_top10 = data_ETL(read_me)
# 剩余假期TOP10
take = data_ETL(read_me,field='剩余调休',reject='天')
rest = data_ETL(read_me,field='剩余假期',reject='天')
read_me['剩余假期2'] = read_me['剩余调休'] + read_me['剩余假期']
rest_top10 = read_me.sort_values(['剩余假期2'],ascending=False).head(10)

# 工作延时次数TOP10
worktimedelay_top10 = data_ETL(read_me,field='工作日延时')
# 司龄TOP10
entertime_top10 = data_ETL(read_me,field='司龄',reject='')
# 缺勤次数TOP10
absence_top10 = data_ETL(read_me,field='缺勤')
# 迟到次数TOP10
belate_top10 = data_ETL(read_me,field='迟到早退')
# 平均工时TOP10
worktimeaverage_top = data_ETL(read_me,field='平均工时',reject='时/天')

read_me[['剩余调休','剩余假期','剩余假期2','姓名']].head()
```

```python
# 创建一个存放 加班 剩余假期 工作延时次数 平均工时 缺勤次数 迟到早退次数
# 以及优秀值的DataFrame 的集合
read_me2 = {}
read_me2['姓名'] = read_me['姓名']
read_me2['加班'] = read_me['公休日加班']
read_me2['剩余假期'] = read_me['剩余假期2']
read_me2['工作延时次数'] = read_me['工作日延时']
read_me2['平均工时'] = read_me['平均工时']
read_me2['缺勤次数'] = read_me['缺勤']
read_me2['迟到早退次数'] = read_me['迟到早退']
read_me2['忙碌值'] = 0.0
# 转换成dataframe类型
read_me2 = pd.DataFrame(read_me2)
```

```python
# 计算公式：y = 剩余假期*0.1+加班*0.2+工作延时次数*0.3+平均工时*0.4 - (缺勤次数+迟到早退次数)*0.5 
read_me2['忙碌值'] = read_me2['剩余假期']*0.1+read_me2['加班']*0.2+read_me2['工作延时次数']*0.3+read_me2['平均工时']*0.4 \
                    - (read_me2['缺勤次数']+read_me2['迟到早退次数'])*0.5
```
