```python
'''
泰坦尼克号获救问题(数据来源于 Kaggle数据集)
    1.一共有1309名乘客数据，其中 891人 是已知存活情况(train.csv),剩下418人 则需要进行分析预测是否存活(test.csv)
    2.字段的意义
        -- 乘客编号[PassengerId]、存活情况(存活：1，死亡：0)[Survived]、客舱等级[Pclass]、乘客姓名[Name]、
            性别[Sex]、年龄[Age]、同乘的兄弟姐妹/配偶数[SibSp]、同乘的父母/小孩数[Parch]、
            船票编号[Ticket]、船票价格[Fare]、客舱号[Cabin]、登船港口[Embarked]
    3.目的：
        -- 通过已知获救的数据，预测乘客生存情况
        
        ----------------****************------------------
        -- 研究的问题1：
            * 整体看来，存活比例如何？
        -- 要求：
            * 读取已知生存数据train.csv
            * 查看已知存活数据中，存活比例如何？
        -- 提示：
            * 注意过程中筛掉缺失值之后 再分析
            * 这里用seaborn制图辅助研究

        
        ----------------****************------------------
        -- 研究的问题2：
            * 结合性别和年龄数据，分析幸存下来的人是那些人？
        -- 要求：
            * 年龄数据的分布情况
            * 男性和女性存活情况
            * 老人和小孩存活情况

        
        ----------------****************------------------
        -- 研究的问题3：
            * 结合SibSp、Parch字段，研究亲人多少与存活的关系
        -- 要求：
            * 有无兄弟姐妹/父母子女和存活与否的关系
            * 亲戚多少与存活与否的关系
            
            
        ----------------****************------------------
        -- 研究的问题4：
            * 结合票的费用情况，研究票价和存活与否的关系
        -- 要求：
            * 票价分布和存活与否的关系
            * 比较研究 生还者和 未生还者的票价情况
        
        
        ----------------****************------------------
        -- 研究的问题5：
            * 利用KNN分类模型，对结果进行预测
        -- 要求：
            * 模型训练字段：Survived,Pclass,Sex,Age,Fare,Family Szie
            * 模型预测test.csv
        -- 提示：
            * 训练数据集中，性别改为数字表示(1：代表男性，0：代表女性)

'''
```

```python
# 导入模块
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import time
# 不发出警告
import warnings
warnings.filterwarnings('ignore')
# 设置工作目录
os.chdir(r'E:\BaiduNetdiskDownload\Titanic-dataset')
```

```python
'''
-- 研究的问题1：
            * 整体看来，存活比例如何？
        -- 要求：
            * 读取已知生存数据train.csv
            * 查看已知存活数据中，存活比例如何？
        -- 提示：
            * 注意过程中筛掉缺失值之后 再分析
            * 这里用seaborn制图辅助研究
'''

train_data = pd.read_csv('train.csv',engine='python')
test_data = pd.read_csv('train.csv',engine='python')

# # 过滤掉空值
# train_data['Survived'].dropna(inplace=True)

# # 计算存活比例
# sur_per = len(train_data[train_data['Survived'] == 1]) / len(train_data) * 100 
# print(' 本次名单中的存活比例约为:%.2f%%' %(sur_per))

# 绘图表示
sns.set_style('ticks')
plt.axis('equal')
train_data['Survived'].value_counts().plot.pie(autopct = '%1.2f%%')
```

```python
'''
 -- 研究的问题2：
            * 结合性别和年龄数据，分析幸存下来的人是那些人？
        -- 要求：
            * 年龄数据的分布情况
            * 男性和女性存活情况
            * 老人和小孩存活情况
'''
# 年龄数据的分布情况
sns.set()
sns.set_style('ticks')

train_data_age = train_data[train_data['Age'].notnull()]

plt.figure(figsize=(12,5))
plt.subplot(121)
train_data_age['Age'].hist(bins=80)
plt.xlabel('Age')
plt.ylabel('Num')

plt.subplot(122)
train_data_age.boxplot(column='Age',showfliers=False)

# 男性和女性存活情况
train_data[['Sex','Survived']].groupby('Sex').mean().plot.bar(figsize=(12,5))
survived_sex = train_data.groupby(['Sex','Survived'])['Survived'].count()
print(' 女性存活率%.2f%%,男性存活率%.2f%%' %
        (survived_sex.loc['female',1] / survived_sex.loc['female'].sum() * 100,
        survived_sex.loc['male',1] / survived_sex.loc['male'].sum() * 100))

train_data_age['Age'].describe()
```

```python
# 分析 年龄与存活率的关系
fig,ax = plt.subplots(1,2,figsize=(18,8))

sns.violinplot('Pclass','Age',hue='Survived',data=train_data_age,split=True,ax=ax[0])
ax[0].set_title('Pclass and Age vs Surviced')

sns.violinplot('Sex','Age',hue='Survived',data=train_data_age,split=True,ax=ax[1])
ax[1].set_title('Sex and Age vs Surviced')

# 查看老人与小孩的存活率
plt.figure(figsize=(18,4))
train_data_age['Age_int'] = train_data_age['Age'].astype(np.int)
average_age = train_data_age[['Age_int','Survived']].groupby('Age_int',as_index=False)
# sns.barplot(x='Age_int',y='Survived',data=average_age,palette='BuPu')
# plt.grid(linestyle='--',alpha=0.5)
```

```python
'''
-- 研究的问题3：
    * 结合SibSp、Parch字段，研究亲人多少与存活的关系
-- 要求：
    * 有无兄弟姐妹/父母子女和存活与否的关系
    * 亲戚多少与存活与否的关系
'''



```

```python
'''
-- 研究的问题5：
    * 利用KNN分类模型，对结果进行预测
-- 要求：
    * 模型训练字段：Survived,Pclass,Sex,Age,Fare,Family Szie
    * 模型预测test.csv
-- 提示：
    * 训练数据集中，性别改为数字表示(1：代表男性，0：代表女性)
'''

train_data['family_size'] = train_data['Parch'] + train_data['SibSp']
knn_train = train_data[['Survived','Pclass','Sex','Age','Fare','family_size']].dropna()
knn_train['Sex'][knn_train['Sex'] == 'male'] = 1
knn_train['Sex'][knn_train['Sex'] == 'female'] = 0

test_data['family_size'] = test_data['Parch'] + test_data['SibSp'] + 1
knn_test = test_data[['Pclass','Sex','Age','Fare','family_size']].dropna()
knn_test['Sex'][knn_test['Sex'] == 'male'] = 1
knn_test['Sex'][knn_test['Sex'] == 'female'] = 0

from sklearn import  neighbors

knn = neighbors.KNeighborsClassifier()
knn.fit(knn_train[['Pclass','Sex','Age','Fare','family_size']],knn_train['Survived'])

knn_test['predict'] = knn.predict(knn_test)
pre_survived = knn_test[knn_test['predict'] == 1].reset_index()
del pre_survived['index']

pre_survived
```
