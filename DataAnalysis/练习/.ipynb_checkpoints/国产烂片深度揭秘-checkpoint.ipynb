```python
'''
 前戏： 导入需要的包
'''
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bokeh.plotting import figure,show,output_file
from bokeh.models import ColumnDataSource,HoverTool
import os

#不发出警告
import warnings
warnings.filterwarnings('ignore')  
```

```python
'''
1.读取数据，以 豆瓣评分 为标准，看看电影评分分布 以及烂片情况(要求)：
    -- 读取数据文件 moviedata.xlsx
    -- 查看 豆瓣评分 数据分布，绘制直方图 和 箱型图
    -- 判断 豆瓣评分 数据是否符合 正太分布
    -- 如果符合正太分布，这里以上四分位(该样本中所有数值由小到大排列后第 25% 的数字) 评分为 烂片标准
    -- 筛选出 烂片数据 并做排名，找到TOP20
    提示：
        * 读取完数据之后，去除缺失值
        * 这里可以用 ks 检验来判断数据是否符合正太分布
'''

os.chdir('xx\xx') # 配置工作目录
df = pd.read_excel('moviedata.xlsx',sheet_name=0)
df = df[df['豆瓣评分'] > 0]

# 豆瓣评分情况 并绘制直方图 、箱型图 进行初步判断
df['豆瓣评分'].describe() # 选取 1/4 位 作为烂片分割线(4.3分)
df['豆瓣评分'].plot.hist(bins=20,color='green',alpha=0.8,figsize=(20,8))

# 进行ks 正太检验
from scipy import stats
u = df['豆瓣评分'].mean()  # 计算均值
std = df['豆瓣评分'].std() # 计算标准差
print(stats.kstest(df['豆瓣评分'],'norm',(u,std))) # 如果计算出来值大于 0.05 则为正太分布

# 选出TOP20 的烂片
data_lp = df[df['豆瓣评分'] < 4.3].reset_index()
lp_TOP20 = data_lp[['电影名称','豆瓣评分','演员','导演']].sort_values(by='豆瓣评分').iloc[:20]
del lp_TOP20['index'] # 删除多余列
```

```python
'''
2.什么题材的电影烂片最多(要求)？
    -- 按照 类型 字段分类，筛选出不同电影属于什么题材
    -- 整理数据 按照 题材 汇总，查看不同题材的烂片比例 并选取TOP20
    -- 将得到的题材烂片比例的TOP20制作散点图(横坐标为 题材 类型，纵坐标为 烂片比例，点大小为样本数量)
        * 用bokeh制图；按照烂片比例做降序排列；
    提示：
        * 删除 类型 字段空值的数据
        * 由于电影类型 会有多个，这里需要将一个电影每个类型都识别出来，在统计每个题材时都需要计算 例如：
            -- 如果一个电影类型为：“喜剧/爱情”，则在计算 喜剧、爱情题材的烂片比例时，都需要将该电影算上
        * 注意类型字段中，要删除空格字符
        * bokeh图设置点大小，这里通过 开放减小数据差距 size = count**0.5*系数
'''

# 筛选出所有题材的类型，查看不同题材烂片的比例
typelist = []
# 选择类型不为空 的进行 去除空格 和 /分割
for i in df[df['类型'].notnull()]['类型'].str.replace(' ','').str.split('/'):
    typelist.extend(i)
    
typelist = list(set(typelist))  # 去除重复 的题材类型
df_type = df[df['类型'].notnull()]  # 进行一次删除空数据的操作

# 自定义一个函数 统计烂片信息(烂片名、烂片个数、烂片占比)
# 注：烂片占比 是指 这种类型的影片(豆瓣评分小于4.3) 占 该类型总量的百分比
# 这样就可以 推论出 那种类型的电影更容易出烂片
def lp_info(data,typei):
    dic_type_lp = {} # 创建空字典
    datai = data[data['类型'].str.contains(typei)] # 筛选属于该题材的数据
    lp_pre_i = len(datai[datai['豆瓣评分']<4.3])/len(datai)
    # 总结输出 烂片信息
    dic_type_lp['typename'] = typei
    dic_type_lp['typecount'] = len(datai)
    dic_type_lp['type_lp_pre'] = lp_pre_i
    
    return dic_type_lp

# 计算 烂片信息
list_type_lp = []
for i in typelist:
    dici = lp_info(df_type,i)
    list_type_lp.append(dici)

# 转换成Dataframe数据类型
df_type_lp = pd.DataFrame(list_type_lp)
# 提取出前20名 烂片类型
type_lp_top20 = df_type_lp.sort_values(by='type_lp_pre',ascending=False).iloc[:20]


# 利用bokeh 制图
type_lp_top20['size'] = type_lp_top20['typecount']**0.5*2  # 设置端的大小
source = ColumnDataSource(type_lp_top20) # 创建绘图数据

list_type = type_lp_top20['typename'].tolist()
hover = HoverTool(tooltips = [('数据量','@typecount'),('烂片率','@type_lp_pre')])

# 设置标签内容
output_file('pic1.html')
p = figure(x_range = list_type,
           plot_width = 900,
           plot_height = 500,
           title = '不同电影题材的烂片比例',
           tools = [hover,'reset,xwheel_zoom,pan,crosshair,box_select'])
p.circle(x='typename',y='type_lp_pre',source=source,size='size',
        line_color='black',line_dash=[6,4],fill_color='red',fill_alpha=0.5)

show(p) # 开始作图
```

```python
'''
3.和什么国家合拍更可能产生烂片(要求)？
    -- 按照制片国家/地区 字段分类，筛选不同电影的制片地
    -- 整理数据 按照题材汇总，查看不同题材的烂片比例
    提示：
        * 删除 制片国家/地区 字段空值的数据
        * 删除 制片国家/地区 中不包括 中国大陆的数据
        * 制片地删除 中国大陆 中国 台湾 香港 等噪音数据
        * 筛选合作电影 大于等于3部以上的国家
'''

df_loc = df[['电影名称','制片国家/地区','豆瓣评分']][df['制片国家/地区'].notnull()]
df_loc = df_loc[df_loc['制片国家/地区'].str.contains('中国大陆')]

loclist = []
for i in df_loc['制片国家/地区'].str.replace(' ','').str.split('/'):
    loclist.extend(i)
loclist = list(set(loclist))
loclist.remove('中国大陆')
loclist.remove('香港')
loclist.remove('台湾')
loclist.remove('中国')

def lp_info2(data,typei):
    dic_loc_lp = {}
    datai = data[data['制片国家/地区'].str.contains(typei)]
    lp_pre_i = len(datai[datai['豆瓣评分']<4.3])/len(datai)
    # 总结输出 烂片信息
    dic_loc_lp['loc'] = typei
    dic_loc_lp['loccount'] = len(datai)
    dic_loc_lp['loc_lp_pre'] = lp_pre_i
    
    return dic_loc_lp

list_loc_lp = []
for i in loclist:
    dici = lp_info2(df_loc,i)
    list_loc_lp.append(dici)

df_loc_lp = pd.DataFrame(list_loc_lp)
df_loc_lp = df_loc_lp[df_loc_lp['loccount']>=3]
loc_lp_top20 = df_loc_lp.sort_values(by = 'loc_lp_pre',ascending = False)

```

```python
'''
4.卡司数量是否和烂片有关(要求)？
    -- 计算每部电影的主演人数
    -- 按照主演人数分类，统计烂片率
        * 分类：1-2人，3-4人，5-6人，7-9人，10人以上
    -- 查看烂片比例最高的演员 TOP20
    提示：
        * 通过 主演 字段内做分列 来计算主演人数
        * 需要分别统计不同主演人数的电影数量以及烂片数量，再计算烂片比例
        * 这里可以按照明星再查看一下 他们的烂片率，比如 吴亦凡、范冰冰、杨幂、黄晓明....
'''

df['主演人数'] = df['主演'].str.replace(' ','').str.split('/').str.len()
df_leaderole01 = df[['主演人数','豆瓣评分']].groupby('主演人数').count()
df_leaderole02 = df[['主演人数','豆瓣评分']][df['豆瓣评分']<4.3].groupby('主演人数').count()

df_leaderole_pre = pd.merge(df_leaderole01,df_leaderole02,left_index=True,right_index=True)
df_leaderole_pre.columns = ['电影数据','烂片数量']

df_leaderole_pre.reset_index(inplace=True)
df_leaderole_pre['主演人数分类'] = pd.cut(df_leaderole_pre['主演人数'],
                                        [0,2,4,6,9,50],
                                        labels =['1-2人','3-4人','5-6人','7-9人','10人以上'])
df_leaderole_pre2 = df_leaderole_pre[['主演人数分类','电影数量','烂片数量']].groupby('主演人数分类').sum()
df_leaderole_pre2['烂片比例'] = df_leaderole_pre2['烂片数量'] / df_leaderole_pre2['电影数量']

```

```python
'''
5.不同导演每年电影产量情况是如何的(要求)？
    -- 通过 上映日期 筛选出每个电影的上映年份
    -- 查看不同导演的烂片比例、这里去除拍过 10次电影以下的导演
    -- 查看不同导演每年的电影产量制作散点图 (横坐标为年份，纵坐标为每年电影平均分，点大小该年电影数量)
        * 用bokeh制图；横坐标限定为2007-2017年；绘制散点图只需要产出烂片的导演数据；
    提示：
        * 注意要删除 上映日期 中的空格字符
        * 绘制图表时，分开建立数据绘制
'''

# 对电影的上映日期进行 整理
df_year = df[['电影名称','导演','豆瓣评分','上映日期']][df['导演'].notnull()] # 筛选导演不为空的
df_year = df_year[df_year['上映日期'].notnull()] # 筛选上映日期不为空的
df_year['上映日期'] = df_year['上映日期'].str.replace(' ','') # 去除空格
df_year['year'] = df_year['上映日期'].str[:4] # 截取年份
df_year = df_year[df_year['year'].str[0]=='2'] # 截取 从 2000 —— 2017年的数据
df_year['year'] = df_year['year'].astype(np.int) # 转换类型

# 筛选导演
directtor_list = []
for i in df_year['导演'].str.replace(' ','').str.split('/'):
    directtor_list.extend(i)
directtor_list = list(set(directtor_list))

# 查看不同导演的烂片比例
list_dir_lp = []

for i in directtor_list:
    datai = df_year[df_year['导演'].str.contains(i)]
    if len(datai) > 10: # 如果这个导演的拍片数大于 10 部再进行统计
        dic_dir_lp = {}
        lp_pre_i = len(datai[datai['豆瓣评分']<4.3])/len(datai)
        
        dic_dir_lp['dir'] = i
        dic_dir_lp['dircount'] = len(datai)
        dic_dir_lp['dir_lp_pre'] = lp_pre_i
        list_dir_lp.append(dic_dir_lp)

df_dir_lp = pd.DataFrame(list_dir_lp)


# 自定义一个函数来 计算不同导演的电影产量 以及 电影均分
def lp_info3(data,diri):
    datai = data[data['导演'].str.contains(diri)]
    data_moviecount = datai[['year','电影名称']].groupby(by='year').count()
    data_scoremean = datai[['year','豆瓣评分']].groupby(by='year').mean()
    df_i = pd.merge(data_moviecount,data_scoremean,left_index=True,right_index=True)
    df_i.columns = ['count','score']
    df_i['size'] = df_i['count'] * 5
    return(df_i)

dirdata1 = lp_info3(df_year,'王晶')

# 导入 BoxAnnotation 模块
from bokeh.models.annotations import BoxAnnotation

# 设置标签的显示内容
hover = HoverTool(tooltips=[('该年电影均分','@score'),
                           ('该年电影产量','@count')])

output_file('pic2.html') # 输出文件
# 构建绘图空间
p = figure(plot_width=900,plot_height=500,title='不同导演每年的电影产量以及电影均分',
              tools=[hover,'reset,xwheel_zoom,pan,crosshair,box_select'])

# 绘制散点图1
source1 = ColumnDataSource(dirdata1)
p.circle(x='year',y='score',source = source1,size='size',legend='王晶',fill_color='green')

# 绘制烂片分割区域
bg = BoxAnnotation(top=4.4,fill_alpha=0.1,fill_color='red')
p.add_layout(bg)

p.xgrid.grid_line_dash = [10,4]
p.ygrid.grid_line_dash = [10,4]
p.legend_location = 'top_right'

show(p)
```
