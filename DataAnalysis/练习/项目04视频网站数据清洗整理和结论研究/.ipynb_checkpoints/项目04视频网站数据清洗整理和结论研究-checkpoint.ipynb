```python
'''
【项目04】  视频网站数据清洗整理和结论研究

作业要求：
1、数据清洗 - 去除空值
要求：创建函数
提示：fillna方法填充缺失数据，注意inplace参数

2、数据清洗 - 时间标签转化
要求：
① 将时间字段改为时间标签
② 创建函数
提示：
需要将中文日期转化为非中文日期，例如 2016年5月24日 → 2016.5.24

3、问题1 分析出不同导演电影的好评率，并筛选出TOP20
要求：
① 计算统计出不同导演的好评率，不要求创建函数
② 通过多系列柱状图，做图表可视化
提示：
① 好评率 = 好评数 / 评分人数
② 可自己设定图表风格

4、问题2 统计分析2001-2016年每年评影人数总量
要求：
① 计算统计出2001-2016年每年评影人数总量，不要求创建函数
② 通过面积图，做图表可视化，分析每年人数总量变化规律
③ 验证是否有异常值（极度异常）
④ 创建函数分析出数据外限最大最小值）
⑤ 筛选查看异常值 → 是否异常值就是每年的热门电影？
提示：
① 通过箱型图验证异常值情况
② 通过quantile(q=0.5)方法，得到四分位数
③ IQR=Q3-Q1
④ 外限：最大值区间Q3+3IQR,最小值区间Q1-3IQR （IQR=Q3-Q1）
⑤ 可自己设定图表风格

'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import warnings 
warnings.filterwarnings('ignore')
```

```python
# 选择文件路径
import os

file_path = r'C:\Python36\Scripts\练习\项目04视频网站数据清洗整理和结论研究'
os.chdir(file_path)

movie_raw_data = pd.read_csv('爱奇艺视频数据.csv',engine='python')

movie_raw_data.head()
```

```python
# 清洗nan数据
def clearning_data(df):
    columns = df.columns
    for col in columns:
        if df[col].dtype == 'object':
            df[col].fillna(value='缺失值',inplace=True)
        else:
            df[col].fillna(value=0,inplace=True)
    return df

movie_data = clearning_data(movie_raw_data)

movie_data.head()
```

```python
'''
2、数据清洗 - 时间标签转化
要求：
① 将时间字段改为时间标签
② 创建函数
提示：
需要将中文日期转化为非中文日期，例如 2016年5月24日 → 2016.5.24
'''

def date_conversion(df,*args):
    for arg in args:
        df[arg] = df[arg].str.replace('年','.')
        df[arg] = df[arg].str.replace('月','.')
        df[arg] = df[arg].str.replace('日','')
    return df
        
data_m = date_conversion(movie_data,'数据获取日期')
data_m.head()
```

```python
'''
3、问题1 分析出不同导演电影的好评率，并筛选出TOP20
要求：
① 计算统计出不同导演的好评率，不要求创建函数
② 通过多系列柱状图，做图表可视化
提示：
① 好评率 = 好评数 / 评分人数
② 可自己设定图表风格
'''

data_comment = data_m.groupby('导演')[['评分人数','好评数']].sum()
data_comment['好评率'] = data_comment['好评数'] / data_comment['评分人数']

nice_comment_rate_TOP20 = data_comment.sort_values(['好评率'],ascending=False)[:20]

# plt.figure(figsize=(30,15))

plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False

plt.xticks(fontsize=18,color='white')
plt.yticks(fontsize=18,color='white') 
plt.title('2001——2016年导演好评率TOP20\n',color='white',fontsize=18)

nice_comment_rate_TOP20['好评率'].plot(
            kind ='bar',
            color ='blue',
            grid = True,
            alpha = 0.5,
            rot = 45,
            ylim = [0.97,1],
            figsize = (30,15)
)
```

```python
'''
问题2 统计分析2001-2016年每年评影人数总量
要求：
① 计算统计出2001-2016年每年评影人数总量，不要求创建函数
② 通过面积图，做图表可视化，分析每年人数总量变化规律
③ 验证是否有异常值（极度异常）
④ 创建函数分析出数据外限最大最小值）
⑤ 筛选查看异常值 → 是否异常值就是每年的热门电影？
提示：
① 通过箱型图验证异常值情况
② 通过quantile(q=0.5)方法，得到四分位数
③ IQR=Q3-Q1
④ 外限：最大值区间Q3+3IQR,最小值区间Q1-3IQR （IQR=Q3-Q1）
⑤ 可自己设定图表风格
'''

data_date = data_m.groupby('上映年份')['评分人数'].sum()
q1 = data_date.quantile(q=0.25)
q3 = data_date.quantile(q=0.75)
iqr = q3 - q1
mi = q1 - 3*iqr
ma = q3 + 3*iqr

# 计算异常值
error = data_date[(data_date < mi) | (data_date > ma)]
# 正常值计算
data_c = data_date[(data_date >= mi) & (data_date <= ma)]

print('异常值共%i条' % len(error))

# 绘制箱型图
fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(2,1,1)
color = dict(boxes='DarkGreen', whiskers='DarkOrange', medians='DarkBlue', caps='Gray')
data_date.plot.box(vert=False,grid=True,color=color,ax=ax1,label='样本数据')

# 绘制散点图
fig.add_subplot(2,1,2)
plt.scatter(data_c.index,data_c,color='g',marker='.',alpha=0.3)
plt.scatter(error.index,error,color='k',marker='.',alpha=0.5)
plt.grid(linestyle='--')

data_date
```

```python
# 答案上的代码
q2data1 = data_m[['导演','上映年份','整理后剧名']].drop_duplicates()
q2data1 = q2data1[q2data1['上映年份'] != 0]

# 求出不同剧的评分人数、好评数总和
q2data2 = data_m.groupby('整理后剧名').sum()[['评分人数','好评数']]
q2data2.head()

# 合并 得到不同年份，不同剧的评分人数、好评数总和
q2data3 = pd.merge(q2data1,q2data2,left_on='整理后剧名',right_index=True,how='inner')
q2data3.head(10)

# 按照电影上映年份统计，评分人数
q2data4 = q2data3.groupby('上映年份').sum()[['评分人数','好评数']]
q2data4.head(10)

# 得到面积图数据
data_area = q2data4['评分人数'].loc[2000:]

data_area.plot.area(figsize=(30,12),grid= True,color='g',alpha=0.8)
plt.xticks(range(2001,2017),fontsize=22,color='white')
plt.yticks(fontsize=22,color='white') 
plt.title('2001——2016年度评分人数量统计',fontsize=22,color='white')
```
