```python
import requests
from lxml import etree

login_tag = requests.Session()
login_tag.auth = ('zhongxinlong','123456')
login_tag.get(r'http://10.28.200.161/zentao/index.php?m=user&f=login')


# 定义一个资源函数 存放请求信息
# 返回 一个列表 和字典



# 定义一个清洗函数 获取禅道首页信息
# 形式如：{'pro_name':'pro_num',.....}
def dq_info(url,path):
    res = login_tag.get(url)
    pro_html = etree.HTML(res.text)
    infos = pro_html.xpath(path)
    data = []
    
    for info in infos:
        data.append(info.text)
    return data
    

def product_info(pro_url,pro_path):
    data = dq_info(pro_url,pro_path)
    dic ={}
    
    for i in range(len(data)-1):
        if i%2 == 0: 
            dic[data[i+1]] = data[i]
    return dic

pro_url = r'http://10.28.200.161/zentao/index.php?m=product&f=all'
pro_path = r'//tbody[@id="productTableList"]//a'
    
pro_dict = product_info(pro_url,pro_path)

num = '29'
bug_url = r'http://10.28.200.161/zentao/index.php?m=bug&f=browse&productID=%s' %(num)
bug_path_1 = r'//table[@id="bugList"]//tr[@class="text-center"]//a'  # 抓取到ID 和 BUG标题
bug_path_2 = r'//table[@id="bugList"]//tr[@class="text-center"]//td[5]'  # 抓取状态
bug_path_3 = r'//table[@id="bugList"]//tr[@class="text-center"]//td[2]/span'  # 抓取严重程度

bug_dict1 = dq_info(bug_url,bug_path_1)
bug_dict2 = dq_info(bug_url,bug_path_2)
bug_dict3 = dq_info(bug_url,bug_path_3)


def bug_info(df1,df2,df3):
    dic = {
        "id":"",
        "title":"",
        "status":"",
        "level":"",
        "sum":"",
    }
    bug_lis =[]
    df1 = list(filter(None,df1))
    i = 1
    while i < len(df1):
        if i%2 != 0: 
            dic["id"] = df1[i-1]
            dic["title"] = df1[i]
            bug_lis.append(dic)
        i += 1
        dic = {"id":"","title":"","status":"","level":"","sum":""}
    
    n=1
    for item in bug_lis:
        item["status"] = df2[n-1]
        item["level"] = df3[n-1]
        item["sum"] = len(df2)
        n += 1
    
    close_num = 0
    nonclose_num = 0
    not_close=[]
    for item in bug_lis:
        if item["status"] == "已关闭":
            close_num += 1
        else:
            nonclose_num += 1
            not_close.append(item)
            
    bug_lis[0]["close_num"] = close_num
    bug_lis[0]["nonclose_num"] = nonclose_num
    
    return bug_lis,not_close

bd = bug_info(bug_dict1,bug_dict2,bug_dict3)

print(bd[0])
print('--------------')
print(bd[1])
```
